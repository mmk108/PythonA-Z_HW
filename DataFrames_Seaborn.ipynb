```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
plt.rcParams['figure.figsize']=10,5

import warnings
warnings.filterwarnings("ignore")
```

```python
vis1 = sns.distplot(stats2["Internet users"])
```

```python
vis1 = sns.distplot(stats2["Internet users"],bins=30)
```

```python
#boxplot

vis2 = sns.boxplot(data=stats2, x="Income Group", y="Birth rate")
```

```python
vis3= sns.lmplot(data=stats2, x="Internet users" , y="Birth rate", fit_reg=False,\
                 hue="Income Group",size =8, aspect =1)
```
